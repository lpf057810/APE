```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor



# Define the health variables and the target variable
health_vars = [
    'bphigh_cru', 'diabetes_c', 'highchol_c', 'CHD_CrudePrev', 
    'DEPRESSION_CrudePrev', 'GHLTH_CrudePrev', 
    'ACCESS2_CrudePrev', 'ARTHRITIS_CrudePrev', 
    'BINGE_CrudePrev', 'BPMED_CrudePrev', 'CANCER_CrudePrev', 
    'CASTHMA_CrudePrev', 'CERVICAL_CrudePrev', 'CHECKUP_CrudePrev', 
    'COLON_SCREEN_CrudePrev', 'COPD_CrudePrev', 
    'COREM_CrudePrev'#,'LPA_CrudePrev','SLEEP_CrudePrev','CHOLSCREEN_CrudePrev','CSMOKING_CrudePrev'
]
target = 'obesity_cr'

# Prepare the data for VIF calculation
X = filtered_data[health_vars].copy()
X = sm.add_constant(X)  # Add a constant column for VIF calculation

# Check for missing or infinite values
print(X.isnull().sum())
print(np.isinf(X).sum())

# Handle missing values by filling them with the median (or any other appropriate value)
X.fillna(X.median(), inplace=True)

# Replace infinite values with large finite numbers
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.fillna(X.median(), inplace=True)

# Calculate VIF values
vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Filter out variables with VIF greater than 20
filtered_vif_data = vif_data[vif_data['VIF'] > 20]

# Remove variables with high VIF from the dataset
filtered_vars = filtered_vif_data['Variable'].tolist()
if 'const' in filtered_vars:
    filtered_vars.remove('const')  # Remove the constant term from the list

# Print the remaining variables and their VIF values
print(filtered_vif_data)

filtered_data=filtered_data.drop(columns=filtered_vars)

```

```python
filtered_data.dropna(subset=['CountyName_x'],inplace=True)
```

```python
# Optimized code with rounding to two decimal places

from sklearn.impute import SimpleImputer
import pandas as pd

# Create a SimpleImputer object to impute missing values using the mean
imputer = SimpleImputer(strategy='mean')

# Impute and round to two decimal places for the selected columns
columns_to_impute = [
    'workers_driving_20_min_to_work', 'workers_public_transport_45_min',
    'percent_population_within_150m_highway(%)', 'normalized_diff_vegetation_index',
    'workers_walking_15_min_to_work', 'Heat Wave - Hazard Type Risk Index Score',
    'Strong Wind - Hazard Type Risk Index Score', 'distance_to_park_within_1_mile'
]

for column in columns_to_impute:
    filtered_data[column] = imputer.fit_transform(filtered_data[[column]]).round(2)

# Proportional imputation for other columns based on population
# Step 1: Calculate the total population
total_population = filtered_data['totalpop2'].sum()

# Step 2: Calculate the median values for imputation
median_values = {
    'Number of Healthcare Facilities': filtered_data['Number of Healthcare Facilities'].median(),
    'Number of Marta Stops': filtered_data['Number of Marta Stops'].median(),
    'INTDENSITY': filtered_data['INTDENSITY'].median(),
    'STRNETDENSITY': filtered_data['STRNETDENSITY'].median(),
    'BLOCKDENSITY': filtered_data['BLOCKDENSITY'].median(),
    'Number of drivers': filtered_data['Number of drivers'].median()
}

# Step 3: Create a proportion column based on population
filtered_data['pop_proportion'] = filtered_data['totalpop2'] / total_population

# Step 4: Impute missing values using the population proportion and round to two decimal places
for column, median_value in median_values.items():
    filtered_data[column] = filtered_data[column].fillna(filtered_data['pop_proportion'] * median_value).round(2)

# Drop the 'pop_proportion' column if it's not needed
filtered_data = filtered_data.drop(columns=['pop_proportion'])
```
